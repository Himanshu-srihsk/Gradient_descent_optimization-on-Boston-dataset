```python
def step_gradient(data,learning_rate,m):
    m_slope = np.zeros(data.shape[1])
    M = data.shape[0]
    N=data.shape[1]
    for j in range(M):
        x=data[j,0:N-1]
        x = np.append(x, 1)
        y=data[j,N-1]
        for i in range(N):
            m_slope[i]+= -(2/M)* (y - (m*x).sum()) * x[i]
              
        m = m - (learning_rate * m_slope)
       
    return m
```

```python
def cost(data,m):
    total_cost = 0
    M=data.shape[0]
    N=data.shape[1]
    for i in range(M):
        x=data[i,0:N-1]
        x = np.append(x, 1)
        y = data[i,N-1]
        total_cost+=(y-(m*x).sum())**2
    return (1/M)*total_cost
```

```python
def gd(data,learning_rate, num_iterations):
    m = np.zeros(data.shape[1])
    for i in range(num_iterations):
        m = step_gradient(data, learning_rate, m)
        if i % 100== 0:
            print(i, " Cost: ", cost(data, m))
    print(i+1, " Cost: ", cost(data, m))        
    return m
```

```python
def predict(x,m,c):
    M=x.shape[0]
    y_pred=np.zeros(M)
    for i in range(M):
        X_data=x[i,:]
        y_pred[i]=(m*X_data).sum()+c
    return y_pred    
    
```
